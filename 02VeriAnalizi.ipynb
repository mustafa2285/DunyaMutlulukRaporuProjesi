# Herhangi bir ülke mutlulukta önemli bir artış veya azalma yaşadı mı?

```python
df2015 = df.loc[df["Year"]== 2015]
```

```python
df2015 =df2015.loc[:,["Country","Happiness Rank"]]
df2015=df2015.rename(columns = {'Happiness Rank':'Happiness_Rank2015'})
```

```python
df2019 = df.loc[df["Year"]== 2019]
```

```python
df2019 =df2019.loc[:,["Country","Happiness Rank"]]
df2019=df2019.rename(columns = {'Happiness Rank':'Happiness_Rank2019'})
```

```python
dfk = pd.merge(df2015,df2019,on="Country")
```

```python
dfk["fark"] = dfk["Happiness_Rank2015"] - dfk["Happiness_Rank2019"]
```

```python
dfk.sort_values("fark",ascending = True)
```

### Mutluluk sıralamasında en fazla artış yaşayan ülke 155'dan 102'ye 53 sıra inen Benin,
### En büyük düşüş ise 23.sıradan 108.sıraya 85 sıra farkla düşen Venezuela  oldu.
